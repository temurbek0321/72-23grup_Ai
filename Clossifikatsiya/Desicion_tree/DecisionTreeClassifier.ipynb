```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Yangi ma'lumotlar
X_new = np.array([[50], [65], [78], [84], [90], [98]])

# Maoshlarni kategoriyalarga ajratamiz:
# 0 - Kichik, 1 - O'rta, 2 - Katta
# (bu faqat misol, real holatda sizga moslab ajrating)
y_new = np.array([0, 0, 1, 1, 2, 2])

# Yangi model (klassifikator)
model_new = DecisionTreeClassifier(max_depth=2)
model_new.fit(X_new, y_new)

# Daraxtni chizish
plt.figure(figsize=(10, 6))
plot_tree(model_new, filled=True, feature_names=["House Size"], class_names=["Kichik", "O'rta", "Katta"], rounded=True)
plt.title("New Decision Tree Classifier")
plt.show()
```
