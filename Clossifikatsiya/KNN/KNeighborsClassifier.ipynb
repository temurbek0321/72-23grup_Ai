```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Ma'lumotlar: [Yosh, Ish tajribasi (yil), Sertifikatlar soni] -> Kasb turi
X = np.array([
    [25, 4, 2],
    [30, 5, 2],
    [29, 4, 1],
    [26, 3, 0],
    [35, 5, 3],
    [32, 4, 2],
    [28, 4, 3],
    [33, 4, 4],
    [27, 3, 1],
    [30, 5, 0],
    [31, 4, 2]
])

y = np.array([0, 2, 0, 0, 2, 0, 1, 1, 1, 2, 0])  # 0: Dasturchi, 1: Dizayner, 2: Menejer

# Model yaratish va o'qitish
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

# Yangi odam uchun bashorat
yangi_odat = np.array([[28, 4, 4]])
pred = model.predict(yangi_odat)[0]

kasblar = {0: "Dasturchi", 1: "Dizayner", 2: "Menejer"}
print(f"Bashorat qilingan kasb: {kasblar[pred]}")

# Modellni baholash
y_pred = model.predict(X)
aniqlik = accuracy_score(y, y_pred)
print(f"Model aniqligi: {aniqlik * 100:.2f}%")

# Batafsil hisobot
print("\nClassification Report:")
print(classification_report(y, y_pred, target_names=["Dasturchi", "Dizayner", "Menejer"]))
```
